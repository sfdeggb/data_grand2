```python
import json
from datetime import datetime
import numpy as np

def calculate_time_decay(current_date, action_date):
    """计算时间衰减权重"""
    days_diff = (datetime.strptime(current_date, '%Y%m%d') - 
                datetime.strptime(action_date, '%Y%m%d')).days
    return np.exp(-0.1 * max(0, days_diff))  # 使用指数衰减

def calculate_category_similarity(item1_first, item1_second, item2_first, item2_second):
    """计算类目相似度"""
    if item1_first == item2_first and item1_second == item2_second:
        return 1.0
    elif item1_first == item2_first:
        return 0.5
    return 0.1

def adjust_scores(inference_results, user_behaviors, current_date='20241205'):
    """
    调整推理分数
    inference_results: 模型推理结果列表 [{'id': id, 'scores': score, ...}, ...]
    user_behaviors: 包含view_seq, clk_seq, order_seq的用户行为字典
    """
    # 行为权重
    behavior_weights = {
        'view': 1.0,   # 浏览权重
        'click': 2.0,  # 点击权重
        'order': 3.0   # 购买权重
    }
    
    # 存储每个商品的累积权重
    item_weights = {}
    
    # 处理所有行为序列
    for behavior_type, seq in [
        ('view', user_behaviors['qysc_view_seq']),
        ('click', user_behaviors['qysc_clk_seq']),
        ('order', user_behaviors['qysc_order_seq'])
    ]:
        if isinstance(seq, str):
            seq = json.loads(seq)
            
        for action in seq:
            if isinstance(action, str):
                action = json.loads(action)
                
            item_id = action['sku_id']
            time_weight = calculate_time_decay(current_date, action['oper_time'])
            
            # 初始化该商品的权重
            if item_id not in item_weights:
                item_weights[item_id] = 0
                
            # 累积权重
            item_weights[item_id] += (behavior_weights[behavior_type] * time_weight)
    
    # 调整推理分数
    adjusted_results = []
    for result in inference_results:
        item_id = result['id']
        original_score = result['scores']
        
        # 如果该商品在用户行为中出现过，调整分数
        if item_id in item_weights:
            adjusted_score = original_score * (1 + item_weights[item_id])
        else:
            adjusted_score = original_score
            
        adjusted_results.append({
            'id': item_id,
            'scores': adjusted_score,
            'original_score': original_score
        })
    
    # 重新归一化分数
    total_score = sum(r['scores'] for r in adjusted_results)
    for result in adjusted_results:
        result['scores'] /= total_score
        
    return adjusted_results

# 使用示例
def test_score_adjustment():
    # 模拟推理结果
    inference_results = [
        {'id': '1001', 'scores': [0.3]},
        {'id': '1002', 'scores': [0.2]},
        {'id': '1003', 'scores': [0.5]}
    ]
    
    # 获取用户行为数据
    sample_data = generate_sample_data()
    user_behaviors = {
        'qysc_view_seq': sample_data['qysc_view_seq'][0],#字符串
        'qysc_clk_seq': sample_data['qysc_clk_seq'][0],
        'qysc_order_seq': sample_data['qysc_order_seq'][0]
    }
    
    # 调整分数
    adjusted_results = adjust_scores(inference_results, user_behaviors)
    print("调整后的结果:", adjusted_results)
```

```python
def adjust_scores(inference_results, user_behaviors, current_date='20241205'):
    """
    调整推理分数
    inference_results: 模型推理结果列表 [{'id': id, 'scores': score, ...}, ...]
    user_behaviors: 包含view_seq, clk_seq, order_seq的用户行为字典
    """
    # 行为权重
    behavior_weights = {
        'view': 1.0,   # 浏览权重
        'click': 2.0,  # 点击权重
        'order': 3.0   # 购买权重
    }
    
    # 存储每个商品的累积权重
    item_weights = {}
    
    # 记录推理结果中已有的商品ID
    existing_item_ids = {result['id'] for result in inference_results}
    
    # 处理所有行为序列
    for behavior_type, seq in [
        ('view', user_behaviors['qysc_view_seq']),
        ('click', user_behaviors['qysc_clk_seq']),
        ('order', user_behaviors['qysc_order_seq'])
    ]:
        if isinstance(seq, str):
            seq = json.loads(seq)
            
        for action in seq:
            if isinstance(action, str):
                action = json.loads(action)
                
            item_id = action['sku_id']
            time_weight = calculate_time_decay(current_date, action['oper_time'])
            
            # 初始化该商品的权重
            if item_id not in item_weights:
                item_weights[item_id] = 0
                
            # 累积权重
            item_weights[item_id] += (behavior_weights[behavior_type] * time_weight)
    
    # 调整推理分数并添加新商品
    adjusted_results = []
    
    # 首先添加用户行为中出现但不在推理结果中的商品
    new_items = []
    for item_id, weight in item_weights.items():
        if item_id not in existing_item_ids:
            # 为新商品设置一个较高的初始分数
            new_items.append({
                'id': item_id,
                'scores': 1.0,  # 设置一个较高的初始分数
                'original_score': 0,
                'is_from_behavior': True  # 标记这是来自用户行为的商品
            })
    
    # 按权重排序新商品
    new_items.sort(key=lambda x: item_weights[x['id']], reverse=True)
    adjusted_results.extend(new_items)
    
    # 然后处理原有的推理结果
    for result in inference_results:
        item_id = result['id']
        original_score = result['scores']
        
        # 如果该商品在用户行为中出现过，调整分数
        if item_id in item_weights:
            adjusted_score = original_score * (1 + item_weights[item_id])
        else:
            adjusted_score = original_score
            
        adjusted_results.append({
            'id': item_id,
            'scores': adjusted_score,
            'original_score': original_score,
            'is_from_behavior': False
        })
    
    # 重新归一化分数
    total_score = sum(r['scores'] for r in adjusted_results)
    for result in adjusted_results:
        result['scores'] /= total_score
        
    return adjusted_results

# 修改测试函数以更好地展示结果
def test_score_adjustment():
    # 模拟推理结果
    inference_results = [
        {'id': '1002', 'scores': [0.2]},
        {'id': '1003', 'scores': [0.5]},
        {'id': '1004', 'scores': [0.3]}
    ]
    
    # 获取用户行为数据
    sample_data = generate_sample_data()
    user_behaviors = {
        'qysc_view_seq': sample_data['qysc_view_seq'][0],
        'qysc_clk_seq': sample_data['qysc_clk_seq'][0],
        'qysc_order_seq': sample_data['qysc_order_seq'][0]
    }
    
    # 调整分数
    adjusted_results = adjust_scores(inference_results, user_behaviors)
    
    print("原始推理结果:")
    for result in inference_results:
        print(f"商品ID: {result['id']}, 分数: {result['scores']:.4f}")
        
    print("\n调整后的结果:")
    for result in adjusted_results:
        source = "来自用户行为" if result.get('is_from_behavior', False) else "来自推理结果"
        print(f"商品ID: {result['id']}, 分数: {result['scores']:.4f}, 来源: {source}")
```

```python
def adjust_scores(inference_results, user_behaviors, sample_data, current_date):
    """
    调整推理分数
    inference_results: 模型推理结果列表 [{'id': id, 'scores': [score], ...}, ...]
    user_behaviors: 包含view_seq, clk_seq, order_seq的用户行为字典
    sample_data: 包含商品类目信息的DataFrame
    current_date: 当前时间
    """
    # 行为权重
    behavior_weights = {
        'view': 1.0,   # 浏览权重
        'click': 2.0,  # 点击权重
        'order': 3.0   # 购买权重
    }
    
    # 创建商品ID到类目信息的映射
    item_category_map = dict(zip(
        sample_data['skuid'],
        zip(sample_data['class_name'], sample_data['goods_class_name'])
    ))
    
    # 存储每个商品的累积权重和类目相似度权重
    item_weights = {}
    category_weights = {}
    
    # 处理所有行为序列
    for behavior_type, seq in [
        ('view', user_behaviors['qysc_view_seq']),
        ('click', user_behaviors['qysc_clk_seq']),
        ('order', user_behaviors['qysc_order_seq'])
    ]:
        if isinstance(seq, str):
            seq = json.loads(seq)
            
        for action in seq:
            if isinstance(action, str):
                action = json.loads(action)
                
            item_id = action['sku_id']
            time_weight = calculate_time_decay(current_date, action['oper_time'])
            
            # 初始化权重
            if item_id not in item_weights:
                item_weights[item_id] = 0
                category_weights[item_id] = 0
            
            # 累积行为权重
            item_weights[item_id] += (behavior_weights[behavior_type] * time_weight)
            
            # 计算类目相似度权重
            for inference_item in inference_results:
                target_id = inference_item['id']
                if target_id in item_category_map:
                    # 获取目标商品的类目信息
                    target_second, target_first = item_category_map[target_id]
                    # 获取行为商品的类目信息
                    action_first = action['frist_class_name']  # 注意原数据中的拼写错误
                    action_second = action['second_class_name']
                    
                    similarity = calculate_category_similarity(
                        action_first, action_second,
                        target_first, target_second
                    )
                    category_weights[target_id] += (similarity * time_weight * behavior_weights[behavior_type])
    
    # 调整推理分数并添加新商品
    adjusted_results = []
    
    # 首先添加用户行为中出现但不在推理结果中的商品
    new_items = []
    for item_id, weight in item_weights.items():
        if item_id not in {r['id'] for r in inference_results}:
            new_items.append({
                'id': item_id,
                'scores': [1.0],  # 设置一个较高的初始分数
                'original_score': [0],
                'is_from_behavior': True
            })
    
    # 按权重排序新商品
    new_items.sort(key=lambda x: item_weights[x['id']], reverse=True)
    adjusted_results.extend(new_items)
    
    # 处理原有的推理结果
    for result in inference_results:
        item_id = result['id']
        original_score = result['scores'][0]  # 获取列表中的分数
        
        # 计算最终权重（结合行为权重和类目相似度权重）
        behavior_weight = item_weights.get(item_id, 0)
        cat_weight = category_weights.get(item_id, 0)
        
        # 调整分数（使用两种权重的加权组合）
        adjusted_score = original_score * (1 + 0.7 * behavior_weight + 0.3 * cat_weight)
        
        adjusted_results.append({
            'id': item_id,
            'scores': [adjusted_score],  # 保持列表格式
            'original_score': [original_score],
            'is_from_behavior': False
        })
    
    # 重新归一化分数
    total_score = sum(r['scores'][0] for r in adjusted_results)
    for result in adjusted_results:
        result['scores'] = [result['scores'][0] / total_score]
        
    return adjusted_results

# 修改测试函数
def test_score_adjustment():
    # 模拟推理结果
    inference_results = [
        {'id': '1001', 'scores': [0.3]},
        {'id': '1002', 'scores': [0.2]},
        {'id': '1003', 'scores': [0.5]}
    ]
    
    # 获取用户行为数据
    sample_data = generate_sample_data()
    user_behaviors = {
        'qysc_view_seq': sample_data['qysc_view_seq'][0],
        'qysc_clk_seq': sample_data['qysc_clk_seq'][0],
        'qysc_order_seq': sample_data['qysc_order_seq'][0]
    }
    
    current_date = '20240115'
    
    # 调整分数
    adjusted_results = adjust_scores(inference_results, user_behaviors, sample_data, current_date)
    
    print("原始推理结果:")
    for result in inference_results:
        print(f"商品ID: {result['id']}, 分数: {result['scores'][0]:.4f}")
        
    print("\n调整后的结果:")
    for result in adjusted_results:
        source = "来自用户行为" if result.get('is_from_behavior', False) else "来自推理结果"
        print(f"商品ID: {result['id']}, 分数: {result['scores'][0]:.4f}, 来源: {source}")
```
